```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apple_data = pd.read_csv('AAPL.csv')
apple_data
```

```python
dates = apple_data['Date']
opens = apple_data['Open']
closes = apple_data['Close']
highs = apple_data['High']
lows = apple_data['Low']
```

```python
plt.plot(dates, closes)
plt.plot(dates, opens)
plt.show()
```

```python
plt.plot(dates, highs)
plt.plot(dates, lows)
plt.show()
```

subplots first two numbers mean the grid in which it will display the graphs and the third number is the index of the graph

```python
fig = plt.figure()
plt.subplot(221)
plt.plot(dates, opens)
plt.subplot(222)
plt.plot(dates, closes)
plt.subplot(223)
plt.plot(dates, highs)
plt.subplot(224)
plt.plot(dates, lows)
plt.show()
```

using axes is good if you know in advance how many graphs you will have

```python
fig, axes = plt.subplots(2, 2)
axes[0,0].plot(dates, opens)
axes[0,1].plot(dates, closes)
axes[1,0].plot(dates, highs)
axes[1,1].plot(dates, lows)
plt.show()
```

## Customizing graphs

```python
plt.plot(dates, closes)
plt.title('Close prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Price in USD')
plt.xticks(pd.date_range('2019-03-20', '2019-04-10', 4).strftime('%Y-%m-%d'))
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
fig.suptitle('Prices for AAPL', y=1.05) # changes location of title
axes[0].plot(dates, highs)
axes[0].set_title('High prices for AAPL')
axes[0].set_xlabel('Dates')
axes[0].set_ylabel('Prices in USD')
axes[1].plot(dates, lows)
axes[1].set_title('Low prices for AAPL')
axes[1].set_xlabel('Dates')
axes[1].set_ylabel('Prices in USD')
plt.tight_layout() # adds separation between two graphs
plt.show()
```

```python
plt.plot(dates, highs)
plt.plot(dates, lows)
plt.title('Daily High and Low prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Price in USD')
plt.xticks(pd.date_range('2019-03-20', '2019-04-10', 4).strftime('%Y-%m-%d'))
plt.legend(loc='upper left') # can also use coordinates -> loc = (0.1, 0.1)
plt.show()
```

```python
plt.plot(dates, closes, 'go--', lw=3) # third paramter changes colour and style of the line (pyplot markers to see more), fourth parameter manages line width
plt.title('Close prices of AAPL')
plt.xlabel('Dates')
plt.ylabel('Price in USD')
plt.xticks(pd.date_range('2019-03-20', '2019-04-10', 4).strftime('%Y-%m-%d'))
plt.show()
```

## Graph types
