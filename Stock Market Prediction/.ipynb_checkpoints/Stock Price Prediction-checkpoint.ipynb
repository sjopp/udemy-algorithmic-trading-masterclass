## Fetching and transforming data

```python
import pandas as pd
```

```python
data = pd.read_csv('AAPL_5_years.csv', header=0, usecols=['Date', 'Close'], parse_dates=True, index_col=['Date'])
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data['Close'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
print(scaled_data)
```

## Creating datasets

```python
train_length = int(len(scaled_data) * 0.7)
test_length = len(scaled_data) - train_length
train_data = scaled_data[0:train_length, :]
test_data = scaled_data[train_length:len(scaled_data), :]
```

```python
import numpy as np
```

```python
# timestep depends on the jump between each data point where 1 would be daily data
def create_dataset(dataset, timestep):
    data_x, data_y = [], []
    for i in range(len(dataset) - timestep - 1):
        data_x.append(dataset[i:(i+timestep), 0])
        data_y.append(dataset[i+timestep, 0])
    return np.array(data_x), np.array(data_y)
```

```python
timestep = 1
train_x, train_y = create_dataset(train_data, timestep)
test_x, test_y = create_dataset(test_data, timestep)

print(train_x)
```

```python
print(train_x.shape)
```

```python
train_x = np.reshape(train_x, (train_x.shape[0], 1, train_x.shape[1]))
test_x = np.reshape(test_x, (test_x.shape[0], 1, test_x.shape[1]))

print(train_x)
```

## Building model

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(256, input_shape=(1,1)))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

## Training and Testing

```python
model.fit(train_x, train_y, epochs=1, batch_size=1, verbose=1)
score = model.evaluate(train_x, train_y, verbose=0)
print('Keras model loss = ', score[0])
print('Keras model accuracy = ', score[1])
```

```python
train_predictions = model.predict(train_x)
test_predictions = model.predict(test_x)

print(train_predictions)
```

```python
train_predictions = scaler.inverse_transform(train_predictions)
train_y = scaler.inverse_transform([train_y])

test_predictions = scaler.inverse_transform(test_predictions)
test_y = scaler.inverse_transform([test_y])

print(train_predictions)
```

```python
train_predict_plot = np.empty_like(scaled_data) # want a matrix with same shape as before
train_predict_plot[:,:] = np.nan # making sure no unwanted values
train_predict_plot[1:len(train_predictions) + 1, :] = train_predictions # inserting our train_predictions correctly formatted
```

```python
test_predict_plot = np.empty_like(scaled_data)
test_predict_plot[:,:] = np.nan
test_predict_plot[len(train_predictions) + 2 + 1: len(scaled_data) - 1, :] = test_predictions
```

```python
plt.plot(scaler.inverse_transform(scaled_data))
plt.plot(train_predict_plot)
plt.plot(test_predict_plot)
plt.show()
```
